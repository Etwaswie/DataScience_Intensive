```python
data = sns.load_dataset("tips")
data.head()
```

## Seaborn

```python
sns.pairplot(data, hue = 'day')
```

```python
sns.pairplot(data, hue = 'sex')
```

### Диаграмма рассеяния

```python
sns.scatterplot(data = data, x = 'total_bill', y = 'tip')
```

```python
sns.scatterplot(data = data, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
sns.scatterplot(data = data, x = 'total_bill', y = 'tip', hue = "smoker")
```

### Линейный график

```python
sns.lineplot(data=data[:50], x = 'total_bill', y = 'tip', hue = 'sex')
```

### Гистограмма распределения

```python
sns.histplot(data['total_bill'], binwidth=0.5, color = 'green')
```

### Ящик с усами

```python
sns.boxplot(data=data, x='day', y='total_bill')
```

```python
sns.boxplot(data=data, x='day', y='total_bill', hue='sex')
```

### Barplot

```python
sns.barplot(data=data, x='size', y='tip')
```

```python
sns.barplot(data=data, x='sex', y='total_bill', hue = 'day')
```

### HeatMap

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot = True)
```

## Matplotlib

```python
plt.scatter(data[data.smoker == "Yes"].total_bill, data[data.smoker == "Yes"].tip)
plt.scatter(data[data.smoker == "No"].total_bill, data[data.smoker == "No"].tip)
plt.xlabel('total_bill')
plt.ylabel('tip')
```

```python
plt.hist(data.total_bill, bins=100);
```

```python
plt.boxplot(data[data.day == 'Sun'].total_bill);
```

```python
plt.imshow(data.corr())
plt.colorbar()
```

```python
alpha = 0.5
plt.figure(figsize=(8, 8))
plt.scatter(data[data.smoker == "Yes"].total_bill, data[data.smoker == "Yes"].tip, alpha=alpha, color = 'g', label = 'Yes')
plt.scatter(data[data.smoker == "No"].total_bill, data[data.smoker == "No"].tip, alpha=alpha, color = 'r', label = 'No')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title("Smoker")

plt.grid()
plt.legend()
```
