## Scatter Plot

```python
plt.figure(figsize=(15,8))
ax = sns.scatterplot(data=data[data['job_title'] == 'Big Data Engineer'], x = 'experience_level', y = 'salary_in_usd')

ax.set_title('Зависимость зарплаты в USD от уровня подготовки Big Data Engineer')
ax.set_ylabel('Зарплата в USD')
ax.set_xlabel('Уровень подготовки Big Data Engineer');
```

```python
plt.figure(figsize=(15,8))
ax1 = sns.scatterplot(data=data, x = 'experience_level', y = 'salary_in_usd')

ax1.set_title('Зависимость зарплаты в USD от уровня подготовки работника')
ax1.set_ylabel('Зарплата в USD')
ax1.set_xlabel('Уровень подготовки работника');
```

```python
#plt.figure(figsize=(20,10))
ax2 = sns.relplot(data=data, x = "salary_in_usd", y = "company_location", hue = "work_year", height = 10, aspect = 1.25);

ax2.fig.suptitle("Зависимость зарплаты в USD от страны по годам", fontsize=12)
ax2.fig.subplots_adjust(top=0.95)
ax2.set_axis_labels(x_var="Зарплата в USD", y_var="Страна");
```

## Line Plot

```python
plt.figure(figsize=(15,8))
ax3 = sns.lineplot(data=data, x = 'experience_level', y = 'salary_in_usd', hue = 'work_year', style = 'work_year', ci = None)

ax3.set_title('Зависимость зарплаты в USD от уровня подготовки работника по годам')
ax3.set_ylabel('Зарплата в USD')
ax3.set_xlabel('Уровень подготовки работника');
```

## Histogram

```python
plt.figure(figsize=(15,8))
ax4 = sns.histplot(data['salary_in_usd'], binwidth=10000)

ax4.set_title('Распределение зарплат в USD')
ax4.set_ylabel('Количество работников')
ax4.set_xlabel('Зарплата в USD');
```

## Диаграмма размаха

```python
plt.figure(figsize=(15,8))
ax5 = sns.boxplot(x = 'experience_level', y = 'salary_in_usd', data = data)

ax5.set_title('Зависимость зарплат в USD от уровня подготовки')
ax5.set_ylabel('Зарплата в USD')
ax5.set_xlabel('Уровень подготовки работника');
```

```python
plt.figure(figsize=(15,8))
ax6 = sns.boxplot(x = 'work_year', y = 'salary_in_usd', data = data, hue = 'experience_level')

ax6.set_title('Зависимость зарплат в USD по годам')
ax6.set_ylabel('Зарплата в USD')
ax6.set_xlabel('Год');
```

## Круговая диаграмма распределения работников по уровням подготовки

```python
plt.figure(figsize=(15,8))

explode = [0, 0, 0, 0.1]
palette_color = sns.color_palette('pastel')

ax7 = plt.pie(data['experience_level'].value_counts().tolist(), labels=data['experience_level'].unique().tolist(),
              autopct='%.0f%%', explode=explode, colors=palette_color);
```

## Круговая диаграмма распределения работников по типу занятости

```python
plt.figure(figsize=(15,8))
explode = [0.1, 0.5, 0.5, 0.5]
plt.pie(data['employment_type'].value_counts().tolist(), labels=data['employment_type'].unique().tolist(), 
        autopct='%.0f%%', explode=explode, colors=palette_color);
```

## Круговая диаграмма распределения компаний по размеру

```python
plt.figure(figsize=(15,8))
plt.pie(data['company_size'].value_counts().tolist(), labels=data['company_size'].unique().tolist(), autopct='%.0f%%'
       , colors=palette_color);
```
